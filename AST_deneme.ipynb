```python
import librosa
import numpy as np
import os
import time

# Giriş ve çıkış yolları
input_paths = {
    "agresif": "/content/drive/MyDrive/Yazgel_Muzik/agresif",
    "huzunlu": "/content/drive/MyDrive/Yazgel_Muzik/huzunlu",
    "neseli": "/content/drive/MyDrive/Yazgel_Muzik/neseli"
}
output_base_path = "/content/drive/MyDrive/Yazgel_Muzik_npy"

# Mel-spectrogram çıkarma fonksiyonu
def extract_mel_spectrogram(audio_path, n_fft=2048, hop_length=512, n_mels=128):
    try:
        y, sr = librosa.load(audio_path, sr=None)  # Ses dosyasını yükle
        spectrogram = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels)
        log_spectrogram = librosa.power_to_db(spectrogram)  # Logaritmik ölçekte spectrogram
        return log_spectrogram
    except Exception as e:
        print(f"HATA: {audio_path} işlenemedi. Sebep: {e}")
        return None

# Çıkış klasörlerini oluştur ve kontrol et
for label in input_paths.keys():
    output_folder = os.path.join(output_base_path, label)
    os.makedirs(output_folder, exist_ok=True)
    print(f"Çıkış klasörü oluşturuldu veya mevcut: {output_folder}")

# İşleme başla
for label, folder_path in input_paths.items():
    output_folder = os.path.join(output_base_path, label)
    print(f"{label} klasöründeki dosyalar işleniyor...")

    if not os.path.exists(folder_path):
        print(f"HATA: {folder_path} klasörü bulunamadı.")
        continue  # Klasör eksikse atla

    start_time = time.time()
    processed_files = set(os.listdir(output_folder))  # Zaten işlenmiş dosyalar

    for i, file_name in enumerate(os.listdir(folder_path)):
        if file_name.endswith('.wav') and file_name.replace('.wav', '.npy') not in processed_files:
            try:
                input_file = os.path.join(folder_path, file_name)
                output_file = os.path.join(output_folder, file_name.replace('.wav', '.npy'))

                mel_spectrogram = extract_mel_spectrogram(input_file)
                if mel_spectrogram is not None:  # Eğer işlem başarılıysa
                    np.save(output_file, mel_spectrogram)
                    elapsed_time = time.time() - start_time
                    print(f"{i+1}. dosya kaydedildi: {output_file} | Geçen süre: {elapsed_time:.2f} saniye")
                else:
                    print(f"HATA: {file_name} işlenemedi. Spectrogram oluşturulamadı.")
            except Exception as e:
                print(f"HATA: {file_name} işlenemedi. Sebep: {e}")
                continue  # Hatalı dosyayı atla
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np
import os

# Veri kümesi sınıfı
class MusicDataset(Dataset):
    def __init__(self, file_paths, label_mapping):
        self.file_paths = []
        self.labels = []
        for label, paths in file_paths.items():
            self.file_paths.extend(paths)
            self.labels.extend([label_mapping[label]] * len(paths))

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        file_path = self.file_paths[idx]
        data = np.load(file_path)  # .npy dosyasını yükle
        data = torch.tensor(data, dtype=torch.float32).unsqueeze(0)  # [1, time_steps, frequency_bins]
        label = self.labels[idx]
        return {"input_values": data, "labels": torch.tensor(label, dtype=torch.long)}

# Çıkış klasörlerinden dosya yollarını ve etiketleri al
npy_paths = {
    "agresif": "/content/drive/MyDrive/Yazgel_Muzik_npy/agresif",
    "huzunlu": "/content/drive/MyDrive/Yazgel_Muzik_npy/huzunlu",
    "neseli": "/content/drive/MyDrive/Yazgel_Muzik_npy/neseli"
}

label_mapping = {"agresif": 0, "huzunlu": 1, "neseli": 2}
file_paths = {label: [os.path.join(path, file) for file in os.listdir(path) if file.endswith('.npy')]
              for label, path in npy_paths.items()}

# Veri kümesini oluştur
dataset = MusicDataset(file_paths, label_mapping)
```

```python
from sklearn.model_selection import train_test_split

# Eğitim ve test setine ayır
train_indices, test_indices = train_test_split(range(len(dataset)), test_size=0.2, random_state=42)

train_dataset = torch.utils.data.Subset(dataset, train_indices)
test_dataset = torch.utils.data.Subset(dataset, test_indices)

print(f"Eğitim seti boyutu: {len(train_dataset)}")
print(f"Test seti boyutu: {len(test_dataset)}")
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
from transformers import AutoModelForAudioClassification, TrainingArguments, Trainer

# AST modelini yükle
model_name = "MIT/ast-finetuned-audioset-10-10-0.4593"
model = AutoModelForAudioClassification.from_pretrained(
    model_name,
    num_labels=len(label_mapping),
    label2id=label_mapping,
    id2label={idx: label for label, idx in label_mapping.items()},
    ignore_mismatched_sizes=True
)

# Eğitim argümanları
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=32,
    num_train_epochs=5,
    weight_decay=0.01,
    save_strategy="epoch",
    logging_dir='./logs',
    load_best_model_at_end=True
)

# Trainer tanımlama
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset
)

# Modeli eğit
print("Model eğitiliyor...")
trainer.train()
```

```python
import numpy as np

# Sabit uzunluk
fixed_length = 1292

def resize_spectrogram(spectrogram, fixed_length):
    current_length = spectrogram.shape[1]
    if current_length < fixed_length:  # Pad with zeros
        pad_width = fixed_length - current_length
        spectrogram = np.pad(spectrogram, ((0, 0), (0, pad_width)), mode='constant')
    elif current_length > fixed_length:  # Crop to fixed length
        spectrogram = spectrogram[:, :fixed_length]
    return spectrogram
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
from transformers import AutoModelForAudioClassification, TrainingArguments, Trainer

model_name = "MIT/ast-finetuned-audioset-10-10-0.4593"
model = AutoModelForAudioClassification.from_pretrained(
    model_name,
    num_labels=len(label_mapping),
    label2id=label_mapping,
    id2label={idx: label for label, idx in label_mapping.items()},
    ignore_mismatched_sizes=True
)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def compute_metrics(pred):
    labels = pred.label_ids
    preds = np.argmax(pred.predictions, axis=1)
    accuracy = accuracy_score(labels, preds)
    precision = precision_score(labels, preds, average='macro')
    recall = recall_score(labels, preds, average='macro')
    f1 = f1_score(labels, preds, average='macro')
    return {"accuracy": accuracy, "precision": precision, "recall": recall, "f1": f1}
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics
)
```

```python
from transformers import DefaultDataCollator

class CustomDataCollator(DefaultDataCollator):
    def __call__(self, features):
        # Giriş ve etiketleri ayır
        input_values = torch.stack([f["input_values"].squeeze(-2) if f["input_values"].dim() == 5 else f["input_values"] for f in features])
        labels = torch.tensor([f["labels"] for f in features])
        return {"input_values": input_values, "labels": labels}
```
