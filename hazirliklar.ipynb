**Etiketleme ve Veri Hazırlığı**

```python
from google.colab import drive
import os
import numpy as np
from sklearn.model_selection import train_test_split

# Veri klasörlerinin yolları; her duygu türü için ayrı klasörler
data_folders = {
    "agresif": "/content/drive/MyDrive/Yazgel_Muzik/npy_dosyalari/agresif.npy",
    "hüzünlü": "/content/drive/MyDrive/Yazgel_Muzik/npy_dosyalari/huzunlu.npy",
    "neşeli": "/content/drive/MyDrive/Yazgel_Muzik/npy_dosyalari/neseli.npy"
}

# Tüm özellik dosyalarını ve etiketlerini bir listeye ekleme
data = []
labels = []
for emotion, folder in data_folders.items():
    for filename in os.listdir(folder):
        if filename.endswith(".npy"):
            filepath = os.path.join(folder, filename)
            data.append(filepath)  # Özellik dosyasının yolu
            labels.append(emotion)  # Dosyaya ait etiket (duygu türü)

# Veriyi eğitim ve test kümelerine ayırma, test kümesi boyutu %10
train_data, test_data, train_labels, test_labels = train_test_split(
    data, labels, test_size=0.1, random_state=42, stratify=labels)

# Eğitim kümesini tekrar eğitim ve validasyon kümelerine ayırma, validasyon boyutu %20
train_data, val_data, train_labels, val_labels = train_test_split(
    train_data, train_labels, test_size=0.2, random_state=42, stratify=train_labels)

# Veri kümelerinin boyutlarını ekrana yazdırma
print(f"Eğitim kümesi: {len(train_data)}")
print(f"Validasyon kümesi: {len(val_data)}")
print(f"Test kümesi: {len(test_data)}")

# Verileri kaydetmek için çıkış klasörü oluşturma
output_folder = "/content/drive/MyDrive/Yazgel_Muzik/BolunmusVeriler"
os.makedirs(output_folder, exist_ok=True)

# Eğitim, validasyon ve test kümelerini .npy dosyası olarak kaydetme
np.save(os.path.join(output_folder, "train_data.npy"), train_data)
np.save(os.path.join(output_folder, "train_labels.npy"), train_labels)
np.save(os.path.join(output_folder, "val_data.npy"), val_data)
np.save(os.path.join(output_folder, "val_labels.npy"), val_labels)
np.save(os.path.join(output_folder, "test_data.npy"), test_data)
np.save(os.path.join(output_folder, "test_labels.npy"), test_labels)

# İşlem tamamlandı mesajı
print("Veri kümeleri başarıyla hazırlandı ve kaydedildi.")
```
